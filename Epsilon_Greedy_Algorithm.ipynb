```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
arms=int(input("Enter no of arms/actions: "))
```

```python
q_star=[]
q_star=np.random.rand(arms)
q_star=q_star/q_star.sum()
print(q_star)
q_star.sum()
```

```python
q_t_action=[]
n_t_a=[]
reward=[]
```

```python

q_t_action=[0]*arms
n_t_a=[0]*arms
reward=[0]*arms
maximum=0
best=[]
epsilon =0.1
cumulative_reward1 = 0
cumulative_rewards_list1 = []
n_iter=100
for i in range(0,100):
    
    num1=np.random.random() 
    if num1< epsilon:
        action=random.choice(range(arms))
    else:
        action = q_t_action.index(max(q_t_action))

    num2=np.random.random()   
    
    
    if num2 < q_star[action]:
        reward[action]=reward[action]+1
        n_t_a[action]=n_t_a[action] + 1
        q_t_action[action]=q_t_action[action]+(reward[action] - q_t_action[action]) / (n_t_a[action] + 1)
        
    else:
        reward[action]=reward[action]+0
        n_t_a[action]=n_t_a[action] + 1
        q_t_action[action]=q_t_action[action]+(reward[action] - q_t_action[action]) / (n_t_a[action] + 1)
    
    cumulative_reward1 += reward[action]
    cumulative_rewards_list1.append(cumulative_reward1)

    print("---------------------------------------------")
    print("TIMESTAMP: ",i)
    print("ACTION CHOOSEN:",action+1)
    print("UPDATED REWARD FOR EACH ACTION",reward)
    print("UPDATED ESTIMATED Q-T:",q_t_action)
    print("TOTAL NUMBER OF CHOOSING EACH ACTIONS",n_t_a)
    if maximum<sum(q_t_action):
        best=q_t_action.copy()
        

print("-----------------------------")
print("BEST_Q_ESTIMATED_VALUE",best)

```

```python
plt.bar(range(len(n_t_a)), n_t_a, tick_label=range(1,len(n_t_a)+1),color='orange')
```
